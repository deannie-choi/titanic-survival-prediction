```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
train_df = pd.read_csv("/Users/dean/Projects/titanic-survival-prediction/datasets/train.csv")
test_df = pd.read_csv("/Users/dean/Projects/titanic-survival-prediction/datasets/test.csv")

# Check missing values
print(train_df.isnull().sum())

# Visualize survival rate by gender
sns.barplot(x="Sex", y="Survived", data=train_df)
plt.title("Survival Rate by Gender")
plt.show()

# Visualize survival rate by Pclass
sns.barplot(x="Pclass", y="Survived", data=train_df)
plt.title("Survival Rate by Ticket Class")
plt.show()
```
